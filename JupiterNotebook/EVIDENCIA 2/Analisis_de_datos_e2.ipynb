2. EXPLORACI칍N GENERAL

```python
# Seleccionar columnas num칠ricas
columnas_numericas = df_copia.select_dtypes(include='number').columns

# Crear una lista para almacenar resultados
resultados = []

for col in columnas_numericas:
    total = len(df_copia)
    validos = df_copia[col].notna().sum()
    nulos = df_copia[col].isna().sum()

    # Se interpretan como mal registrados o err칩neos los valores negativos o cero.
    erroneos = df_copia[col][df_copia[col] <= 0].count()

    media = df_copia[col].mean()
    mediana = df_copia[col].median()
    std = df_copia[col].std()

    resultados.append({
        "Columna": col,
        "Total": total,
        "V치lidos": validos,
        "Nulos": nulos,
        "Err칩neos": erroneos,
        "Media": round(media, 2),
        "Mediana": round(mediana, 2),
        "Desv칤o est치ndar": round(std, 2)
    })

# Mostrar resultados
pd.DataFrame(resultados)
```

## An치lisis de columnas categ칩ricas (v치lidos, nulos, mal registrados y cantidad de registros por categor칤a)

1. EXPLORACI칍N GENERAL

```python
# Seleccionar solo las columnas categ칩ricas (no num칠ricas)
columnas_categoricas = df_copia.select_dtypes(exclude='number').columns

# Guardar resultados en una lista
resultados = []

for col in columnas_categoricas:
    total = len(df_copia)
    validos = df_copia[col].notna().sum()
    nulos = df_copia[col].isna().sum()
    categorias = df_copia[col].nunique()

    # Se interpretan como mal registrados o err칩neos los valores vac칤os
    erroneos = df_copia[col].isin(["", " ", "-", "Sin dato", "No aplica"]).sum()

    resultados.append({
        "Columna": col,
        "Total registros": total,
        "V치lidos": validos,
        "Nulos": nulos,
        "Mal registrados (vac칤os o inv치lidos)": erroneos,
        "Cantidad de categor칤as distintas": categorias
    })

# Mostrar resultados
pd.DataFrame(resultados)
```
